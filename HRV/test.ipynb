```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {'Name': ['John', 'Alice', 'Bob', 'Emily'],
        'Age': [25, 32, 28, 36]}
df = pd.DataFrame(data)

# Create a figure and axis
fig, ax = plt.subplots()

# Hide axis
ax.axis('off')

# Create a table on the figure and display the DataFrame
table = plt.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

# Set the table properties
table.auto_set_font_size(False)
table.set_fontsize(14)
table.scale(1.2, 1.2)

# Show the figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for demonstration
data = np.random.randn(1000)  # Random normally distributed data

# Calculate statistical measures
mean = np.mean(data)
std = np.std(data)
min_val = np.min(data)
max_val = np.max(data)

# Create a figure
fig, ax = plt.subplots()

# Plot a histogram
ax.hist(data, bins=30, density=True, alpha=0.7, color='blue')

# Add statistical report as text
report = f"Mean: {mean:.2f}\nStd: {std:.2f}\nMin: {min_val:.2f}\nMax: {max_val:.2f}"
ax.text(0.7, 0.9, report, transform=ax.transAxes, fontsize=10, verticalalignment='top')

# Add labels and title
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Histogram with Statistical Report')

# Show the figure
plt.show()
```
